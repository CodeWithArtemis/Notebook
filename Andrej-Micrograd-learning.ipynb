```python
class value:
    def __init__(self,data):
        self.data=data
    def __repr__(self):
        return f"Value(data={self.data})"
    def __add__(self,other):
        result=value(self.data+other.data)
        return result
    def __mul__(self,other):
        result=value(self.data * other.data)
        return result
    def __sub__(self,other):
        result=value(self.data-other.data)
        return result
a=value(2)
b=value(-3)
c=value(3)
a*b+c
```
